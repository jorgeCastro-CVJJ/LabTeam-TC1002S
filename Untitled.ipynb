# HeatMaps

Aquí tenemos las representaciones de los coeficientes de pearson para deducir que variables tienen o no una correlación. De esta manera podemos enfocar nuestro análisis a esos datos.  

```python
colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sb.heatmap(df.corr(), linewidths=0.1, vmax=1.0, square=True, 
           cmap=colormap, linecolor='white', annot=True);
```
